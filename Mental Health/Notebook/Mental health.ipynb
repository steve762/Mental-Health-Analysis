```python
#DATA CLEANING
#removing redudant columns
columns_to_drop=['state','Timestamp','comments']
data.drop(columns=columns_to_drop, inplace=True)
```

```python
data
```

```python
#having overview of age column 
data['Age'].describe()
```

```python
#checking age column data type
data['Age'].dtype
```

```python
# removing outliers
data=data[(data['Age']>=18) & (data['Age']<=65)].copy()
```

```python
data['Age'].describe()
```

```python
#CLEANING GENDER COLUMN
#changing everything to lowercase and strip extra spaces
data.loc[:,'Gender']=data['Gender'].str.lower().str.strip()
```

```python
data
```

```python
#Grouping gender to male, female and other
def clean_gender (g):
    if g in('male','Androgyne','Cis male', 'Cis man', 'M', 'maile','Mail', 'Male(CIS)'):
        return 'Male'
    elif g in('Female','Cis Female', 'cis-female/femme', 'femail','Femake', 'Female(cis)','woman','queer/she/they'):
        return 'Female'
    else: 
        return 'Other'
data['Gender']=data['Gender'].apply(clean_gender)
```

```python
data['Gender'].unique()
```

```python
#cleaning self_employed column
data['self_employed']=data['self_employed'].astype(str).str.strip()
data['self_employed']=data['self_employed'].replace('nan','Unknown')
```

```python
data['self_employed'].unique()
```

```python
data['work_interfere']=data['work_interfere'].astype(str).str.strip()
data['work_interfere']=data['work_interfere'].replace('nan','Unknown')
```

```python
data['work_interfere'].unique()
```

```python
#cleaning no_employees column
data['no_employees'].unique()
```

```python
data['no_employees'] = data['no_employees'].astype(str).str.strip()
data['no_employees']=data['no_employees'].replace({'1-5':'1-5','6-25':'6-25',
                                                  '26-100':'26-100','100-500':'100-500',
                                                  '500-1000':'500-1000','More than 1000':'1000+',
                                                  '01-May':'1-5', 'Jun-25':'6-25'})
data['no_employees']=" " + data['no_employees']
```

```python
data['no_employees'].unique()
```

```python
#cleaning other columns
cols = ['benefits', 'care_options', 'wellness_program', 'seek_help', 
        'mental_health_consequence', 'phys_health_consequence',
        'coworkers', 'supervisor', 'mental_vs_physical', 'leave']

for col in cols:
    data[col] = data[col].replace({"Don't know":"Unknown","Not sure":"Unknown",
                                   "Maybe":"Unknown","Some of them":"Unknown"})
```

```python
#Export cleaned dataset
data.to_csv("/Users/user/Downloads/cleaned_survey2.csv", index=False)
```
